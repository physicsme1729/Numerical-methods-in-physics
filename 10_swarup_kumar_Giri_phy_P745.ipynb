# Question 1

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *

def Lennard_potetial(x,eplison=1):
    return 4*eplison*(x**12 - x**6)
x = np.linspace(0.5,1,1000)
y=[Lennard_potetial(i) for i in x ]
plt.ylim([-1.1,0])
plt.plot(x,y)
plt.xlabel("1/r")
plt.ylabel("V")
plt.show()
```

```python
import math
def bracket(f,x1,h):
  c = 1.618033989
  f1 = f(x1)
  x2 = x1 + h; f2 = f(x2)
  # Determine downhill direction and change sign of h if needed
  if f2 > f1:
    h = -h
    x2 = x1 + h; f2 = f(x2)
# Check if minimum between x1 - h and x1 + h
    if f2 > f1: return x2,x1 - h
# Search loop
  for i in range (100):
    h = c*h
    x3 = x2 + h; f3 = f(x3)
    if f3 > f2: return x1,x3
    x1 = x2; x2 = x3
    f1 = f2; f2 = f3
  print("Bracket did not find a minimum")
def search(f,a,b,tol=1.0e-9):
  nIter = int(math.ceil(-2.078087*math.log(tol/abs(b-a))))
  R = 0.618033989
  C = 1.0 - R
# First telescoping
  x2 = C*a + R*b; x1 = R*a + C*b
  f1 = f(x1); f2 = f(x2)
# Main loop
  for i in range(nIter):
    if f1 > f2:
      a = x1
      x1 = x2; f1 = f2
      x2 = C*a + R*b; f2 = f(x2)
    else:
      b = x2
      x2 = x1; f2 = f1
      x1 = R*a + C*b; f1 = f(x1)
  if f1 < f2: return x1,f1
  else: return x2,f2

xStart = 0.96
h = 0.01
x1,x2 = bracket(Lennard_potetial,xStart,h)
x,fMin = search(Lennard_potetial,x1,x2)
print("x =",x)
print("V(x) =",fMin)
```

Solving it analytically


Taking $\epsilon$ =1 and $\sigma$=1   


$F(x) = 4[x^{(12)} - x^{(6)}]$


To find minima
$$ F'(x) = 0 \\
\implies 12x^{11} - 6x^{5} = 0 \\
\implies 2x^{6} - 1 = 0 \\
\implies x_{min}= \left(\frac{1}{2}\right)^{\frac{1}{6}} = 0.89089 $$


$$ F(x_{min}) = -1$$
This matches with result we obtained above
