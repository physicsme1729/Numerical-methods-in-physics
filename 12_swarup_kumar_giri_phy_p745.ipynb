# ***Problem-1:***

Linear regression can be extended to data that depend on two or more variables
(called multiple linear regression). If the dependent variable is z and independent variables are x and y, the data to be fitted have the form

|      |              |           |
|------|--------------|-----------|
| x1   | y1           | z1        |
| x2   | y2           | z2        |
| x3   | y3           | z3        |
| x4   | y4           | z4        |
| ---- | ----         | ----      |
| ---- | ----         |-----      |
| xn   | yn           | zn        |

Instead of a straight line, the fitting function now represents a plane:

f(x, y) = a + bx + cy

Show that the normal equations for the coefficients are

$\begin{bmatrix}
n & \sum{x_i} & \sum {y_i}& \\
\sum{x_i} &\sum{x_i^2} & \sum{x_iy_i} \\
\sum {y_i} & \sum{x_iy_i} &\sum{y_i^2} \\
\end{bmatrix}
\begin{bmatrix}
a \\
b\\
c\\
\end{bmatrix}$
=
$\begin{bmatrix}
\sum {z_i}\\
\sum{x_iz_i} \\
\sum{y_iz_i}
\end{bmatrix}$


# ***Ans:-***

To derive the normal equations for the coefficients of a multiple linear regression model, we start with the equation for the plane:

$
f(x, y) = a + bx + cy
$

We have data points in the form of (x, y, z). The goal is to find the values of the coefficients a, b, and c that minimize the sum of squared differences between the observed z values and the values predicted by the plane.

The sum of squared differences (the cost or error function) is given by:


$E = \sum_{i=1}^{n}(z_i - f(x_i, y_i))^2$

Now, let's differentiate \(E\) with respect to each of the coefficients a, b, and c and set the derivatives equal to zero. This will give us the normal equations. Here's how you can do it:

1. Differentiate E with respect to a:


$\frac{dE}{da} = -2\sum_{i=1}^{n}(z_i - (a + bx_i + cy_i))$

Setting this derivative to zero:

$-2\sum_{i=1}^{n}(z_i - (a + bx_i + cy_i)) = 0$

Dividing by -2:


$\sum_{i=1}^{n}(z_i - (a + bx_i + cy_i)) = 0$

Now, let's rearrange this equation:


$\sum_{i=1}^{n}z_i - \sum_{i=1}^{n}(a + bx_i + cy_i) = 0$


$\sum_{i=1}^{n}z_i - a\sum_{i=1}^{n} - b\sum_{i=1}^{n}x_i - c\sum_{i=1}^{n}y_i = 0$

2. Differentiate E with respect to b:


$\frac{dE}{db} = -2\sum_{i=1}^{n}(z_i - (a + bx_i + cy_i))x_i$

Setting this derivative to zero:


$-2\sum_{i=1}^{n}(z_i - (a + bx_i + cy_i))x_i = 0$

Dividing by -2:


$\sum_{i=1}^{n}(z_i - (a + bx_i + cy_i))x_i = 0$

Now, let's rearrange this equation:


$\sum_{i=1}^{n}z_ix_i - a\sum_{i=1}^{n}x_i - b\sum_{i=1}^{n}x_i^2 - c\sum_{i=1}^{n}x_iy_i = 0$

3. Differentiate E with respect to c:


$\frac{dE}{dc} = -2\sum_{i=1}^{n}(z_i - (a + bx_i + cy_i))y_i$

Setting this derivative to zero:


$-2\sum_{i=1}^{n}(z_i - (a + bx_i + cy_i))y_i = 0$
Dividing by -2:


$\sum_{i=1}^{n}(z_i - (a + bx_i + cy_i))y_i = 0$

Now, let's rearrange this equation:


$\sum_{i=1}^{n}z_iy_i - a\sum_{i=1}^{n}y_i - b\sum_{i=1}^{n}x_iy_i - c\sum_{i=1}^{n}y_i^2 = 0$

Now, we have three equations for a, b, and c. Rearrange them as a matrix equation:

$\begin{bmatrix}
n & \sum{x_i} & \sum {y_i}& \\
\sum{x_i} &\sum{x_i^2} & \sum{x_iy_i} \\
\sum {y_i} & \sum{x_iy_i} &\sum{y_i^2} \\
\end{bmatrix}
\begin{bmatrix}
a \\
b\\
c\\
\end{bmatrix}$
=
$\begin{bmatrix}
\sum {z_i}\\
\sum{x_iz_i} \\
\sum{y_iz_i}
\end{bmatrix}$


These are the normal equations for the coefficients a, b, and c in the multiple linear regression model. The solution to this system of equations will give us the values of a, b, and c that minimize the sum of squared differences between the observed and predicted z values.

# ***Problem -3:***

Debye’s formula for the heat capacity $C_V$ as a solid is $C_V$ = 9Nkg(u), where

\begin{equation}
    g(u)=u^3\int_{0}^{\frac{1}{u}} \frac{x^4e^x}{(e^x-1)^2} \,dx
\end{equation}

The terms in this equation are
                      
N = number of particles in the solid

k = Boltzmann constant

$u=\frac{T}{\Theta_D}$

T = absolute temperature

$\Theta_D$=Debye temperature

Compute g(u) from u = 0 to 1.0 in intervals of 0.05 and plot the results.



# ***Ans:-***

I use  romberg code to plotting this graph

```python
import numpy as np
import matplotlib.pyplot as plt

def richardson(Rprev, Rincurr0, i):
    Rcurr = [Rincurr0]
    for j in range(1, i + 1):
        val = (4**j * Rcurr[j - 1] - Rprev[j - 1]) / (4**j - 1)
        Rcurr.append(val)
    return Rcurr

def trapezoid(f, a, b, n):
    h = (b - a) / (n - 1)
    xs = a + np.arange(n) * h
    cs = np.ones(n)
    cs[0] = 0.5
    cs[-1] = 0.5
    contribs = cs * f(xs)

    contribs = np.where(np.isfinite(contribs), contribs, 0)
    return h * np.sum(contribs)


def prettyprint(row):
    for elem in row:
        print("{0:1.10f} ".format(elem), end="")
    print("")

def romberg(f, a, b, imax=20, tol=1.e-8):
    n = 2
    val = trapezoid(f, a, b, n)
    Rprev = [val]
    prettyprint(Rprev)
    for i in range(1, imax):
        nprime = 2 * n - 1
        Rincurr0 = trapezoid(f, a, b, nprime)
        Rcurr = richardson(Rprev, Rincurr0, i)
        prettyprint(Rcurr)
        err = abs(Rprev[-1] - Rcurr[-1]) / abs(Rcurr[-1])
        valprime = Rcurr[-1]
        if err < tol:
            return valprime
        n = nprime
        Rprev = Rcurr[:]
    else:
        return None
    return valprime


def integrand(x, u):
    return x**4 * np.exp(x) / (np.exp(x) - 1)**2

def g(u):
    result = u**3 * romberg(lambda x: integrand(x, u), 0, 1/u, tol=1.e-8)
    return result

# Initialize arrays to store u and g(u) values
u_values = np.arange(0.05, 1.05, 0.05)
g_values = []

# Calculate g(u) for each u value
for u in u_values:
    g_u = g(u)
    g_values.append(g_u)

# Plot the results
plt.plot(u_values, g_values, marker='o', label='g(u)')
plt.xlabel('u')
plt.ylabel('g(u)')
plt.title('Plot of g(u) vs. u')
plt.grid(True)
plt.legend()
plt.show()
```
